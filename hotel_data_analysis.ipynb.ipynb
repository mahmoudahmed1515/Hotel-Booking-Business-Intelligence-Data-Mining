## 🌟 SECTION E — FINAL EDA SUMMARY

### Q20 | Top 5 Insights

1. **HIGH CANCELLATION RATE (37.1% — €16.7M lost revenue)**
   Over 1 in 3 bookings is canceled. The estimated lost revenue represents a massive operational and financial risk for hotel management.
2. **LEAD TIME IS THE STRONGEST PREDICTOR**

3. **THE NON-REFUND DEPOSIT ANOMALY**
   Non-Refund bookings show a 99.4% cancellation rate. Investigation reveals near-zero special requests, very long lead times, and below-average ADR, suggesting bulk OTA inventory or system artifacts.
4. **CITY HOTEL IS SIGNIFICANTLY MORE VOLATILE**
   City hotels face a 41.7% cancellation rate compared to 27.7% for Resort hotels, driven by Transient online bookers.
5. **CUSTOMER PROFILE MATTERS MORE THAN PRICE**
   Special requests, parking needs, and booking changes are stronger cancellation predictors than ADR. Committed guests leave behavioral signals.

### 🛠 Top 5 Data Problems Identified
1. `company` column: 94.3% missing → dropped.
